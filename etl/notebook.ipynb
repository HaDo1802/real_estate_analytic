````python
import pandas as pd

# Extract the list of properties from the data dictionary
properties = data.get('props', [])
df = pd.DataFrame(properties)

# Drop columns with all null values
df_cleaned = df.dropna(axis=1, how='all')

# Drop rows with missing essential fields (e.g., address or price)
essential_fields = ['address', 'price']
df_cleaned = df_cleaned.dropna(subset=essential_fields)

# Reset index after cleaning
df_cleaned = df_cleaned.reset_index(drop=True)

df_cleaned.head()
````
